```python
import random

class FalseLLM:
    def __init__(self):
        print('LLM created')
    def predict(self,prompt):
        response_list=['Delhi is the capital of india','virat kohli is a cricket player',
                      'Narendra Modi is prime minister of India']
        return {'response':random.choice(response_list)}
```

```python
llm=FalseLLM()
```

```python
llm.predict("What is the capital of India")
```
